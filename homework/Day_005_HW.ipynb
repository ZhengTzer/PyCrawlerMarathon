```python
import requests
import json

# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
popularPost = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
popData = json.loads(popularPost.text)
print(len(popData))
for x in popData[0]:
    print(x)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for i in popData:
    print('æ¨™é¡Œ: ', i['title'],
          ' \nè²¼æ–‡æ™‚é–“: ', i['createdAt'],
          ' \nç•™è¨€äººæ•¸: ', i['commentCount'],
          ' \næŒ‰è®šäººæ•¸: ', i['likeCount'], '\n')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
popCom = 0
popLike = 0

npCom = 0
npLike = 0

for i in popData:
    popCom = popCom + i['commentCount']
    popLike = popLike + i['likeCount']
```

```python
notPopPost = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
notPopData = json.loads(notPopPost.text)

for i in notPopData:
    npCom = npCom + i['commentCount']
    npLike = npLike + i['likeCount']
```
