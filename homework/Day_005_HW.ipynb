## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
import json

# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
popularPost = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
popData = json.loads(popularPost.text)
print(len(popData))
for x in popData[0]:
    print(x)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for i in popData:
    print('標題: ', i['title'],
          ' \n貼文時間: ', i['createdAt'],
          ' \n留言人數: ', i['commentCount'],
          ' \n按讚人數: ', i['likeCount'], '\n')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
popCom = 0
popLike = 0

npCom = 0
npLike = 0

for i in popData:
    popCom = popCom + i['commentCount']
    popLike = popLike + i['likeCount']
```

```python
notPopPost = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
notPopData = json.loads(notPopPost.text)

for i in notPopData:
    npCom = npCom + i['commentCount']
    npLike = npLike + i['likeCount']
```

```python
print('熱門文章:\n平均留言人數: ', round(popCom/len(popData),2),
        '\n平均按讚人數: ', round(popLike/len(popData),2),
      
      '\n\n最新文章:\n平均留言人數: ', round(npCom/len(notPopData),2),
      '\n平均按讚人數: ', round(npLike/len(notPopData),2))
```
